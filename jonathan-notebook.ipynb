```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://api.open-meteo.com/v1/forecast?latitude=-22.91&longitude=-43.18&hourly=temperature_2m'

city = {'Rio' :{'lat':'-22.91', 'lon':'-43.18'},
       'Paris':{'lat':'48.85', 'lon':'2.35'}
       }

#meteo in Rio
time = requests.get(url).json()['hourly']['time']
temperature = requests.get(url).json()['hourly']['temperature_2m']
```
